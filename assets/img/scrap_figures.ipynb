```python
        #Generate plot object
fig,ax=plt.subplots(figsize=(6,6))

#Plot polynomial of order 2
##Capture range
x=np.linspace(-100,100,100)
##Define function values at locations of interest
x_p2=np.array([-100,0,100])
y_p2=np.array([0,11000,0])
##Get numpy to do the heavy lifting of fitting the polynomial
pfit2=np.polyfit(x_p2,y_p2,2)
##Instantiate function to return values from that fitted polynomial
pval2=np.poly1d(pfit2)
##Plot poly
y2=[pval2(val) for val in x]
ax.plot(x,y2,lw=4)

#Plot polynomial of order 4
##Define function values at locations of interest
x_p4=np.array([-100,-50,0,70,100])
y_p4=np.array([0,11000,7000,8000,0])
##Get numpy to do the heavy lifting of fitting the polynomial
pfit4=np.polyfit(x_p4,y_p4,4)
##Instantiate function to return values from that fitted polynomial
pval4=np.poly1d(pfit4)
##Plot poly
y4=[pval4(val) for val in x]
ax.plot(x,y4,c='r',lw=4)

#Provide some shading
ax.fill_between(x,[pval2(val) for val in x],alpha=.3)
ax.fill_between(x,[pval4(val) for val in x],alpha=.3,color='r')

#Fix ranges
ax.set_xlim([-100,100])
ax.set_ylim([0,12000])

#Set labels
ax.set_title('Comparison of Utility Profiles for 1-D Goods')
ax.set_xlabel('x')
ax.set_ylabel('u=f(x)')

#Fix axes
simpleaxis(ax)

plt.savefig('utility_1d_goods_compare.svg')

print pfit2
print pfit4
```
