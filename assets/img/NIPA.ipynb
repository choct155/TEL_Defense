We will also need recession bars to place trajectories in context.  For this purpose we have pulled a time series of binary, NBER-based, [recession indicators](http://research.stlouisfed.org/fred2/series/USREC#) from the [FRED](http://research.stlouisfed.org/fred2/) repository hosted by the St. Louis Fed branch.

```python
#Read in data
rec=pd.read_csv(data_dir+'nber_recessions_fred.csv',parse_dates='observation_date')

#Convert date to datetime
rec['date']=rec['observation_date'].apply(lambda x: pd.to_datetime(x))

#Extract year and month
rec['year']=rec['date'].apply(lambda x: x.year)
rec['month']=rec['date'].apply(lambda x: x.month)

#Calculate month fraction and add to year for starts and stops
rec['stop']=(rec['month']/12.)+rec['year']
rec['start']=(rec['month']/12.-(1/12.))+rec['year']

rec.head()
```

## Intergovernmental Transfers to Local Governments

What about transfers to localities?  I grabbed some NIPA data on summary levels of transfers.

```python
#Establish data location
md_dir='/home/choct155/dissertation/MiscData/'

#Define columns names
col_names=['line','source_tmp']+range(1959,2014)

#Read in data
intgov=pd.read_csv(md_dir+'loc_intgov.csv',skiprows=5,nrows=43,header=None,names=col_names)

#Fix source names
intgov['source']=intgov['source_tmp'].apply(lambda x: x.strip())

#Set index
intgov.set_index('source',inplace=True)

intgov
```

Let's isolate intergovernmental transfers, and deflate them via PCE.

```python
#Subset
ig=intgov.ix[['Federal grants-in-aid','State government grants-in-aid']]

#Drop unnecessary columns
ig.pop('line')
ig.pop('source_tmp')

#Transpose
igt=ig.T.astype(float)

#Fix column names
igt.columns=['fed_transfer','st_transfer']

#Incorporate PCE
igt['pce']=nipa['pce'].ix[igt.index].values

#Generate deflator
igt['pce_defl']=igt['pce'].apply(lambda x: x/100.)

#Capture inflation adjusted transfers
igt['fed_igt']=igt['fed_transfer']/igt['pce_defl']
igt['st_igt']=igt['st_transfer']/igt['pce_defl']

igt.head(10)
```
